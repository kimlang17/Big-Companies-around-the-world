## Scraping data from weather website

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_="wikitable sortable")
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df = df.rename(columns={'Revenue (USD millions)': 'revenue'})
```

```python
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
# check revenue datatype
df['revenue'].dtype
```

```python
# change it from string to float
df['revenue'] = df['revenue'].str.replace(',', '').astype(float)
```

```python
#check revenue data type again
df['revenue'].dtype
```

```python
# Sort the DataFrame by 'Revenue' in descending order
df = df.sort_values(by='revenue', ascending=False)

# Select the top ten rows
top_ten = df.head(10)

# Plotting the bar graph
plt.bar(top_ten['Name'], top_ten['revenue'], color='blue')
plt.xlabel('Products')
plt.ylabel('revenue')
plt.title('Top Ten Products by Revenue')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df['Employees'].dtype
```

```python
# change it from string to float
df['Employees'] = df['Employees'].str.replace(',', '').astype(float)
```

```python
df['Employees'].dtype
```

```python
# Sort the DataFrame by 'Employees' in descending order
df = df.sort_values(by='Employees', ascending=False)

# Select the top ten rows
top_ten = df.head(10)

# Plotting the bar graph
plt.bar(top_ten['Name'], top_ten['Employees'], color='blue')
plt.xlabel('Products')
plt.ylabel('Employees')
plt.title('Top Ten Products by Employees')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Get the top 10 most frequent values in the 'Industry' column
top_10_industries = df['Industry'].value_counts().head(10)

# Plot a pie chart for the top 10 industries
plt.pie(top_10_industries, labels=top_10_industries.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Industries by Frequency')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
# change it from string to float
df['Revenue growth'] = df['Revenue growth'].str.replace('%', '').astype(float)
```

```python
# Sort the DataFrame by 'Revenue' in descending order
df = df.sort_values(by='Revenue growth', ascending=False)

# Select the top ten rows
top_ten = df.head(10)

# Plotting the bar graph
plt.bar(top_ten['Name'], top_ten['Revenue growth'], color='blue')
plt.xlabel('Products')
plt.ylabel('Revenue Growth')
plt.title('Top Ten Products by Revenue')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
